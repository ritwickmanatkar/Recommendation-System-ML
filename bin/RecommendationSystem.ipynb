```python
import os
from xml.dom import minidom
import pandas as pd
import numpy as np
import sklearn
```

```python
current_directory = str(os.getcwd())
path, directory_name = os.path.split(current_directory)
config_file = os.path.join(str(path),"cfg\config.xml")
print(config_file)

parsed_file = minidom.parse(config_file)
data = parsed_file.getElementsByTagName("path")
data_path = data[0].attributes['data_path'].value
print(data_path)
```

```python
user_file = os.path.join(str(data_path),'u.user')
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv(user_file, sep='|', names=u_cols,
                    encoding='latin-1')
users.head(5)
```

```python
data_file = os.path.join(str(data_path),'u.data')
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv(data_file, sep='\t', names=r_cols,
                      encoding='latin-1')
ratings.head(5)
print(max(ratings['unix_timestamp']))
```
