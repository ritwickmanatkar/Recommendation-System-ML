```python
def month_convert(month):
    if month == "Jan":
        return 1
    elif month == "Feb":
        return 2
    elif month == "Mar":
        return 3
    elif month == "Apr":
        return 4
    elif month == "May":
        return 5
    elif month == "Jun":
        return 6
    elif month == "Jul":
        return 7
    elif month == "Aug":
        return 8
    elif month == "Sep":
        return 9
    elif month == "Oct":
        return 10
    elif month == "Nov":
        return 11
    elif month == "Dec":
        return 12
    else:
        return -1
```

```python
import os
from xml.dom import minidom
import pandas as pd
import numpy as np
import sklearn
import time
from datetime import date
```

```python
current_directory = str(os.getcwd())
path, directory_name = os.path.split(current_directory)
config_file = os.path.join(str(path),"cfg\config.xml")
print(config_file)

parsed_file = minidom.parse(config_file)
data = parsed_file.getElementsByTagName("path")
data_path = data[0].attributes['data_path'].value
print(data_path)
```

```python
user_file = os.path.join(str(data_path),'u.user')
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv(user_file, sep='|', names=u_cols, encoding='latin-1')
users.head(5)
```

```python
data_file = os.path.join(str(data_path),'u.data')
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv(data_file, sep='\t', names=r_cols,encoding='latin-1')
ratings.head(5)
print(ratings[ratings['unix_timestamp'] == max(ratings['unix_timestamp'])])

current_unix_timestamp = max(ratings['unix_timestamp']) + 9999

import datetime
current_timestamp = datetime.datetime.fromtimestamp(current_unix_timestamp)
print(current_timestamp)
```

```python
from datetime import datetime
movie_file = os.path.join(str(data_path),'u.item')
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv(movie_file, sep='|', names=i_cols,
 encoding='latin-1')

items['timestamp'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
print(items.head())
items = items[items["movie title"] != "unknown"]


for index, row in items.iterrows():
    day , month , year = row['release date'].split("-")
    datetime_object = datetime.strptime(str(month + " "+ day + " " +year) , '%b %d %Y')
    items.at[index,'timestamp'] = datetime_object
    
print(items.head())
```

```python
# Simple Popularity Model
items['total ratings'] = 0
items['avg ratings'] = 0
for index, row in ratings.iterrows():
    try:
        i = items.index[items['movie id'] == row["movie_id"]].tolist()
        print(index ,i[0])
    except:
        continue
    items.at[i[0] , 'total ratings'] += 1
    items.at[i[0] , 'avg ratings'] += int(row["rating"])
items['avg ratings'] /= items['total ratings']

print(items[["movie id", "avg ratings" , "total ratings"]])
```

```python
m = items['total ratings'].quantile(0.90)
print(m)
```

```python
simple = items[items['total ratings'] >= m]
print(simple.shape)

c = items['avg ratings'].mean()
print(c)
```

```python
def weighted_rating(x,m=m,c= c):
    v = x['total ratings']
    r = x['avg ratings']
    return (v/(v+m)*r)+(m/(m+v)*c)
```

```python
simple['score'] = simple.apply(weighted_rating,axis=1)

simple = simple.sort_values('score',ascending=False)

simple[['movie title','total ratings', 'avg ratings', 'score']].head(15)
```
